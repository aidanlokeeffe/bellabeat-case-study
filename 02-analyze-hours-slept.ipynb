```python
import pandas as pd
```

```python
# Load in the minute-by-minute sleep data
minute_sleep = pd.read_csv("fitabase-data-4.12.16-5.12.16/minute_sleep.csv")
minute_sleep["date"] = pd.to_datetime(minute_sleep["date"])
minute_sleep["rowId"] = minute_sleep["Id"].astype("string") + "-" + minute_sleep["logId"].astype("string")
```

```python
# Iterate over the row Ids to get the duration of sleep sessions
rowIds = pd.unique(minute_sleep["rowId"])
data = []
for rowId in rowIds:
    filt = (minute_sleep["rowId"] == rowId)
    sleep_session = minute_sleep.loc[filt]

    Id_parts = rowId.split("-")

    Id = Id_parts[0]
    start = min(sleep_session["date"])
    end = max(sleep_session["date"])
    elapsed = end - start
    logId = Id_parts[1]

    data.append([Id, start, end, elapsed, logId])

    del filt

sleep_data = pd.DataFrame(data)

```

```python
# Clean up the sleep data
sleep_data.columns = ["Id", "starttime", "endtime", "duration", "logId"]

buffer = sleep_data["duration"].astype("string").str.split(pat=" ", expand=True)
sleep_data["duration"] = buffer[2]
```
